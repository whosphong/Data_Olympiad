### 1st filtering by removing buildings with more than 40% NaN values

```python
mask = (df.isna().sum() / len(df)) <= 0.4
filtered_df = df.loc[:, mask]
filtered_df
```

```python
def drop_columns_with_24h_window(df, n):
    valid_cols = []

    for col in df.columns:
        # Rolling window of 24 hours to count non-nan values
        valid_counts = df[col].notna().rolling(window=24, min_periods=24).sum()
        if (valid_counts[23:] >= n).all():
            valid_cols.append(col)

    return df[valid_cols]
```

### 2nd filtering by removing buildings with less than 2 records in 24h rolling window

```python
filtered_df2 = drop_columns_with_24h_window(filtered_df, 2)
```
